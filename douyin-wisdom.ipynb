```python
import jieba
import pandas as pd

with open('Documents/Posts-wisdom.txt', 'r', encoding='utf-8') as file:
    data = file.readlines()

seg_list = [list(jieba.cut(line.strip(), cut_all=False)) for line in data]
```

```python
stopwords = [line.strip() for line in open("Documents/chinese_stopwords.txt", 'r', encoding='utf-8').readlines()]

processed_docs = [[word for word in doc if word not in stopwords] for doc in seg_list]
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(processed_docs)
dictionary.filter_extremes(no_below=15, no_above=1, keep_n=100000)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
from gensim.models import LdaMulticore
lda_model = LdaMulticore(bow_corpus, num_topics=10, id2word=dictionary, passes=2, workers=2)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, bow_corpus, dictionary)
pyLDAvis.display(vis)
```
